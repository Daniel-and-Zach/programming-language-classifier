```python

def find_percent_word(feature_list, dataframe):
    for feature in feature_list:
        def temp_fn(word_list):
            counter = 0
            for word in word_list:
                if word == feature:
                    counter += 1
            return counter/len(word_list)
        dataframe[feature] = dataframe["Textblob"].apply(temp_fn)
        
        
def find_if_word(feature_list, dataframe):
    for feature in feature_list:
        def temp_fn(word_list):
            for word in word_list:
                if word == feature:
                    return 1
            return 0
        dataframe[feature] = dataframe["Textblob"].apply(temp_fn)
        
        
def find_percent_character(feature_list, dataframe):
    for feature in feature_list:
        def temp_fn(character_list):
            counter = 0
            for character in character_list:
                if character == feature:
                    counter += 1
            return counter/len(character_list)
        dataframe[feature] = dataframe["Textblob letters"].apply(temp_fn)
        
        

def find_percent_special_chars(special_chars, dataframe):
    for feature in special_chars:
        def temp_fn(text):
            counter = len(re.findall(feature, text))
            return counter/len(text.split())
        dataframe[feature] = dataframe["Text"].apply(temp_fn)
```

```python
common_words =['Object', 'var', 'try', 'except', 'class', 'return', 
               'self', 'public', 'static', 'Body', 'val', 'defn', 
               'def', 'end', 'each', 'let', 'define', 'printf',
               "function", "echo", "global", "foreach", "elif",
               "data", "move", "where"]
common_chars = ["!", "(", ")", ";", ":", "$", "@","[", "]", "|", "{", "}"]
double_chars = ["=>", "->", "php"]
```

```python

def check_freqs(language):
   lang = df.Language == language
   lang = df[lang]
   all_txt = lang.Textblob.sum(axis = 0)
   list_words = []
   for word in all_txt:
       if len(word) > 2:
           list_words.append(word)        
   return (Counter(list_words).most_common(5))
    
def char_freqs(language):
    lang = df.Language == language
    lang = df[lang]
    all_text = df["Text"].sum(axis = 0)
    #print(all_text)
    list_char = re.findall(r"[^A-Za-z0-9\n\\ ]", all_text)
    return Counter(list_char)

        
def make_feature_set(list_of_langs):
    feature_set = set()
    for lang in list_of_langs:
        lang_words = check_freqs(lang)
        for word_tuple in lang_words:
            feature_set.add(word_tuple[0])
    return feature_set


#def make_unique_features(list_of_langs):
#    for lang in list_of_langs:
#        lang_words = check_freqs(lang)
#        for word_tuple in lang_words:
#            feature_set.add(word_tuple[0])
#    return feature_set
```

```python
attributes_train, attributes_test, class_train, class_test = train_test_split(df.loc[0::,'Object':"->"], df.loc[0::,"Language"], test_size=0.4, random_state=0)

from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier = classifier.fit(attributes_train, class_train)
predicted = classifier.predict(attributes_test)

from sklearn import metrics
print(metrics.classification_report(class_test, predicted))
print(metrics.confusion_matrix(class_test, predicted))
print(metrics.f1_score(class_test, predicted))

from sklearn.cross_validation import cross_val_score
```

```python
language_tree = DecisionTreeClassifier(max_depth=10)
language_tree = language_tree.fit(attributes_train, class_train)
predict_tree = language_tree.predict(attributes_test)
print(metrics.classification_report(class_test, predict_tree))
print(metrics.confusion_matrix(class_test, predict_tree))
print(metrics.f1_score(class_test, predict_tree))
```

```python
def page_scraper(scrape_page, name_of_file):
    lang_dict = {"clojure": ".clojure", "haskell" : ".hs", "java": ".java",
                 "javascript": ".javascript", "ocaml": ".ocaml", "perl" : ".pl",
                 "php" : ".php", "python" : ".py", "ruby" : ".rb",
                 "scala" : ".scala", "scheme" : ".scm"}
    file_dir = "training/benchmarks/benchmarksgame/bench/"
    page = requests.get(scrape_page)
    start_soup = BeautifulSoup(page.text)
    #print(start_soup)
    for lang in lang_dict.keys():
        lang_with_html = BeautifulSoup(str(start_soup("pre", {"class" : lang + " highlighted_source"})))
        code_text = lang_with_html.get_text()
        code_text = ("").join(code_text)
        if code_text == "[]":
            pass
        else:
            parse_text = code_text[1:len(code_text)-1]
            temp_file = open(file_dir + name_of_file + lang_dict[lang], "+w")
            temp_file.write(parse_text)
            temp_file.close()
```
